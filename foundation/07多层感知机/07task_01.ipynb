```python
import pandas as pd
import numpy as np
data = pd.read_csv('data.csv')
data.head()
```

```python
X = data.drop(['y'],axis=1)
y = data.loc[:,'y']
X.head()
```

```python
#visualize the data
%matplotlib inline
from matplotlib import pyplot as plt
fig1 = plt.figure(figsize=(5,5))
passed=plt.scatter(X.loc[:,'x1'][y==1],X.loc[:,'x2'][y==1])
failed=plt.scatter(X.loc[:,'x1'][y==0],X.loc[:,'x2'][y==0])
plt.legend((passed,failed),('passed','failed'))
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('raw data')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=10)
print(X_train.shape,X_test.shape,X.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense,Activation

mlp=Sequential()
mlp.add(Dense(units=20,input_dim=2,activation='sigmoid')) #第一个隐层，含有20个单元，输入为2，激活函数sigmoid
mlp.add(Dense(units=1,activation='sigmoid'))  #输出层，因为是二分类，故输出只有一个单元，激活函数sigmoid
mlp.summary()   #查看神经网络各层
```

```python
mlp.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
mlp.fit(X_train,y_train,epochs=3000)
```

```python
y_train_predict=mlp.predict_classes(X_train)
from sklearn.metrics import accuracy_score
acc_train=accuracy_score(y_train,y_train_predict)
print(acc_train)
```

```python
y_test_predict = mlp.predict_classes(X_test)
accuracy_test = accuracy_score(y_test,y_test_predict)
print(accuracy_test)
```
